b)Function that finds k closest points to a given point.

**closest** : this function takes input as the new points(x_new,y_new) and the data set generated previously.It calculates the distance of each point in the numpy array and returns a sorted list(which is a pair list and is of the form : (distance,(x,y)).

The distance is calculated using **linalg.norm** which returns the norm of vector or matrix.

```
def closest(x,y,coordinates):
  closest_points = []
  new_point = np.array((x,y))
  for p in coordinates :
    dis = np.linalg.norm(new_point-p)
    closest_points.append((dis,p))
  closest_points.sort()
  return closest_points
```

**k_closest** : this function takes input as k(number of points) and the pair list that would have been generated by the previous function(closest).This function takes the first k points and stores them in a list.The list is then returned.If the number of points in the array are less than k,the we return the remaining points and print a warning.

```
def k_closest(k,pair):
  if (len(pair) < k):
    print("Less than k points present in the input array")
    print("So returning the remaining points")
    k = len(pair)
  k_closest_points = []
  for i in range(k):
    k_closest_points.append(pair[i][1])
  return k_closest_points
```

Let the input point be (1,1).We find the 5,10 and 15 closest points uisng k_closest function and store them in their respective lists.

```
x = int(input("Enter x_new : "))
y = int(input("Enter y_new : "))
dis_and_points = closest(x,y,coordinates)
k = int(input(("Enter k : ")))
arr_5 = k_closest(k,dis_and_points)
print("Done...")
k = int(input(("Enter k : ")))
arr_10 = k_closest(k,dis_and_points)
print("Done...")
k = int(input(("Enter k : ")))
arr_15 = k_closest(k,dis_and_points)
print("Done...")
```

c)Function that generates k points that make a positive angle to a given point.

**pos_angle** : This function takes the input point and coordinates as its input.It finds the points that make positive angle from the numpy array and stores them in a list.The list is then returned.

Positive angle is confirmed by checking if **np.dot** of the two points is positive.  

```
def pos_angle(x,y,coordinates):
  temp = []
  new_point = np.array((x,y))
  for p in coordinates:
    if np.dot(new_point,p) >= 0 :
        temp.append(p);
  return temp
```

**k_posang** : this function takes input as k(number of points) and the points that would have been generated by the previous function(pos_angle).This function takes the first k points and stores them in a list.The list is then returned.
If the number of points in the array are less than k,then we return the remaining points in the array and print a warning.

```
def k_posang(k,points):
  if (len(points) < k):
    print("Less than k points present in the input array")
    print("So returning the remaining points")
    k = len(points)
  k_posang_points = []
  for i in range(k):
    k_posang_points.append(points[i])
  return k_posang_points
```

```
posangle_coordinates = pos_angle(x,y,coordinates)
k = int(input(("Enter k : ")))
arr_5 = k_posang(k,posangle_coordinates)
print("Done...")
k = int(input(("Enter k : ")))
arr_10 = k_posang(k,posangle_coordinates)
print("Done...")
k = int(input(("Enter k : ")))
arr_15 = k_posang(k,posangle_coordinates)
print("Done...")
```

###**Q4.Plotting graphs of various functions**###

In the below code,we generate 100 random points in the interval [-0.5,0.5] and plot the sample mean given by y = (x1+x2+....xk)/k.We also plot the functions (1/k)^(1/2) and -(1/k)^(1/2).

```
x = np.random.uniform(-0.5,0.5,size = 100)
sum_xi = np.zeros(100)
k = np.zeros(100)
for i in range(100):
  if i != 0 :
    sum_xi[i] += sum_xi[i-1] + x[i]
  else : 
    sum_xi[i] += x[i]
  k[i] = i + 1
```

```
plt.plot(k,sum_xi/k,color = 'r',label = 'y')
plt.plot(k,(1/k)**(1/2),color = 'b',label = 'f1')
plt.plot(k,-1*((1/k)**(1/2)),color = 'g',label = 'f2')
plt.title("Plots of funcitons y,f1 and f2")
plt.xlabel('k')
plt.ylabel('Y-axis')
plt.legend()
```
