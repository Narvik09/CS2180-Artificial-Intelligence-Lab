Importing Libraries.

```python
import numpy as np
import pandas as pd
```

Initializing input arrays.

```python
action = np.array([1,2,3,4,6])#shot array
p_out_min = np.array([0.01,0.02,0.03,0.1,0.3])#probability of top batsmen getting out
p_out_max = np.array([0.1,0.2,0.3,0.5,0.7])#probability of last batsmen getting out
```

Probability calculating function.

```python
def p_out(w,index):
  return p_out_max[index] + ((p_out_min[index] - p_out_max[index])*(w-1)/9)
```

Computing the best score and best shot.

```python
num_balls = 300
num_wickets = 10
#initializing 2D numpy arrays for storing best score and best shot at each ball.
best_score = np.zeros((num_balls + 1,num_wickets + 1))
best_shot = np.zeros((num_balls + 1,num_wickets + 1))

for b in range(1,301):
  for w in range(1,11):
    for a in range(5):
      shot = action[a]#shot for a particular ball
      score = (1-p_out(w,a))*(shot + best_score[b-1][w]) + p_out(w,a)*(best_score[b-1][w-1])#computing the score based on probabilities.
      #storing the best score and best shot.
      if (score > best_score[b][w]):
        best_score[b][w] = score
        best_shot[b][w] = shot
```

Best Score Array



```python
print(pd.DataFrame(np.floor(best_score)))#printing the floor of the predicted score.
```

Best Shot Array

```python
print(pd.DataFrame(best_shot))
```
